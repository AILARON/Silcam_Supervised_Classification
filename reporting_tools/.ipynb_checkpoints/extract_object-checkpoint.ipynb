```python
depths = edges_all #[1.00160781, 4.91043308, 8.81925835, 12.72808362, 16.63690889, 20.54573416,
         # 24.45455943, 28.3633847, 32.27220997, 36.18103524, 40.08986051]
object_concentration = []
for i in range(0,len(depths)-1):
    #print(depths[i+1] - depths[i])
    stat_small_all = stat[(stat['depth'] >= depths[i]) & (stat['depth'] < depths[i+1])]
    print("#################################")
    #df[c].unique()
    num_of_objects = stat_small_all["timestamp"].unique().size
    print(stat_small_all["timestamp"].unique().size)
    print("#################################")
    print('all objects ', stat_small_all['density'].count(), stat_small_all['density'].min(), stat_small_all['density'].max())
    stat_small_ch = stat_copepod[(stat_copepod['depth'] >= depths[i]) & (stat_copepod['depth'] < depths[i+1])]
    print('copepods highest prob ', stat_small_ch['density'].count(), stat_small_ch['density'].min(), stat_small_ch['density'].max())
    print('copepods highest prob sample_volume: ', stat_small_ch['density'].count()/ (num_of_objects * sample_volume))
    stat_small_c95 = stat_copepod_95[(stat_copepod_95['depth'] >= depths[i]) & (stat_copepod_95['depth'] < depths[i+1])]
    print('95 copepod ', stat_small_c95['density'].count(), stat_small_c95['density'].min(), stat_small_c95['density'].max())
    print('copepods 95 sample_volume: ', stat_small_c95['density'].count()/ (num_of_objects * sample_volume))

    stat_small_dh = stat_diatom_chain[(stat_diatom_chain['depth'] >= depths[i]) & (stat_diatom_chain['depth'] < depths[i+1])]
    print('diatom chains highest prob ', stat_small_dh['density'].count(), stat_small_dh['density'].min(), stat_small_dh['density'].max())
    print('diatom chains highest prob sample_volume: ', stat_small_dh['density'].count()/ (num_of_objects * sample_volume))
    stat_small_d95 = stat_diatom_chain_95[(stat_diatom_chain_95['depth'] >= depths[i]) & (stat_diatom_chain_95['depth'] < depths[i+1])]
    print('95 diatom chain ', stat_small_d95['density'].count(), stat_small_d95['density'].min(), stat_small_d95['density'].max())
    print('diatom chain 95 sample_volume: ', stat_small_d95['density'].count()/ (num_of_objects * sample_volume))
    
    stat_small_fh = stat_faecal_pellets[(stat_faecal_pellets['depth'] >= depths[i]) & (stat_faecal_pellets['depth'] < depths[i+1])]
    print('faecal pellets highest prob ', stat_small_fh['density'].count(), stat_small_fh['density'].min(), stat_small_fh['density'].max())
    print('faecal pellets highest prob sample_volume: ', stat_small_fh['density'].count()/ (num_of_objects * sample_volume))
    stat_small_f95 = stat_faecal_pellets_95[(stat_faecal_pellets_95['depth'] >= depths[i]) & (stat_faecal_pellets_95['depth'] < depths[i+1])]
    print('95 faecal pellets ', stat_small_f95['density'].count(), stat_small_f95['density'].min(), stat_small_f95['density'].max())
    print('faecal pellets 95 sample_volume: ', stat_small_f95['density'].count()/ (num_of_objects * sample_volume))

    
    object_concentration.append(
        {
            "min depth" : depths[i], 
            "max depth": depths[i+1],
            "depth": int(depths[i] + (depths[i+1]-depths[i])/2), #round(depths[i] + (depths[i+1]-depths[i])/2, 2),
            "min density":min(stat_small_all['density'].min(), 
                              stat_small_ch['density'].min(),
                              stat_small_c95['density'].min(),
                              stat_small_dh['density'].min(),
                              stat_small_d95['density'].min(),
                              stat_small_fh['density'].min(),
                              stat_small_f95['density'].min()), 
            "max density":max(stat_small_all['density'].max(), 
                              stat_small_ch['density'].max(),
                              stat_small_c95['density'].max(),
                              stat_small_dh['density'].max(),
                              stat_small_d95['density'].max(),
                              stat_small_fh['density'].max(),
                              stat_small_f95['density'].max()), 
            "count all": num_of_objects, 
            "min density all": stat_small_all['density'].min(), 
            "max density all": stat_small_all['density'].max(),
            
            "count copepod highest prob": stat_small_ch['density'].count(), 
            "sample volume copepod highest prob": stat_small_ch['density'].count()/ (num_of_objects * sample_volume),    
            "min density copepod highest prob": stat_small_ch['density'].min(), 
            "max density copepod highest prob": stat_small_ch['density'].max(),
            "count copepod 95": stat_small_c95['density'].count(), 
            "sample volume copepod 95": stat_small_c95['density'].count()/ (num_of_objects * sample_volume),    
            "min density copepod 95": stat_small_c95['density'].min(), 
            "max density copepod 95": stat_small_c95['density'].max(),
            
            "count diatom chains highest prob": stat_small_dh['density'].count(), 
            "sample volume diatom chains highest prob": stat_small_dh['density'].count()/ (num_of_objects * sample_volume),    
            "min density diatom chains highest prob": stat_small_dh['density'].min(), 
            "max density diatom chains highest prob": stat_small_dh['density'].max(),
            "count diatom chains 95": stat_small_d95['density'].count(), 
            "sample volume diatom chains 95": stat_small_d95['density'].count()/ (num_of_objects * sample_volume),    
            "min density diatom chains 95": stat_small_d95['density'].min(), 
            "max density diatom chains 95": stat_small_d95['density'].max(),
            
            "count faecal pellets highest prob": stat_small_fh['density'].count(), 
            "sample volume faecal pellets highest prob": stat_small_fh['density'].count()/ (num_of_objects * sample_volume),    
            "min density faecal pellets highest prob": stat_small_fh['density'].min(), 
            "max density faecal pellets highest prob": stat_small_fh['density'].max(),
            "count faecal pellets 95": stat_small_f95['density'].count(), 
            "sample volume faecal pellets 95": stat_small_f95['density'].count()/ (num_of_objects * sample_volume),    
            "min density faecal pellets 95": stat_small_f95['density'].min(), 
            "max density faecal pellets 95": stat_small_f95['density'].max(),
        }
    )

object_concentration = pd.DataFrame(object_concentration)
print(object_concentration.shape)
print(object_concentration)
```

```python
### create montages
## Make montages of processed particle images
maxlength = 60000 #5000000
minlength = 1200
msize = 200 # 2048

def montager(stats1, dmin, dmax, feature, cmin, cmax):  
    '''
    Wrapper for pysilcam montage maker
    '''
    stats = stats1
    
    stats = stats[~np.isnan(stats['major_axis_length'])]
    stats = stats[(stats['major_axis_length'] *
            settings.PostProcess.pix_size) < maxlength]
    stats = stats[(stats['major_axis_length'] *
            settings.PostProcess.pix_size) > minlength]
    
    stats = stats[stats['depth'] >= dmin] 
    stats = stats[stats['depth'] <= dmax]
    stats = stats[stats[feature] >= cmin] 
    stats = stats[stats[feature] <= cmax]

    stats.sort_values(by=['major_axis_length'], ascending=False, inplace=True)
    roifiles = scpp.gen_roifiles(stats, auto_scaler=msize)

    montage = scpp.montage_maker(roifiles, roidir, settings.PostProcess.pix_size, msize, eyecandy=True)
    return montage

def silcam_montage_plot(montage, settings):
    scplt.montage_plot(montage, settings.PostProcess.pix_size)
    
def summary_figure(montage, settings, detobj, depths, density):
    fig = plt.figure(figsize=(500, 250))
    fig, axs = plt.subplots(1,int(len(montage)))
    
    for i in range(0, int(len(montage))):
        plt.sca(axs[i])
        min_depth = '{:0.0f}'.format(depths[i][0])
        max_depth = '{:0.0f}'.format(depths[i][1])
        if density[i][0] <= 1025:
            min_density = '{:0.0f}-'.format(density[i][0])
            max_density = '{:0.0f}'.format(density[i][1])
        else:
            min_density = 'greater than '
            max_density = '{:0.0f}'.format(density[i][1])
        
        #min_depth +'m'+ ' < depth < ' + max_depth+'m'
        axs[i].set_title('depth ' + min_depth + '-' + max_depth + ' meters', fontsize=10)
                         # +'\n' +
        axs[i].set_ylabel('density ' + min_density + max_density + 'kg/m\u00b3') # , fontsize=14    
                         
                        #min_density +'kg/m\u00b3'+ ' < density < ' + max_density+'kg/m\u00b3')
        scplt.montage_plot(montage[i], settings.PostProcess.pix_size)
    plt.tight_layout()
    fig.savefig(OUTPUTDIR + '/' + detobj + '_depth_' + min_depth + '_' + max_depth
                + '_montage.png', dpi=600, bbox_inches='tight')
    plt.close(fig) 
```
